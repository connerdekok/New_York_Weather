```python
import pandas as pd
```

```python
# Load the CSV files into DataFrames
hourly_df = pd.read_csv("../Albany_New_York_Weather/data/hourly_data.csv")
# Display the first few rows from each file to understand their structure
print("Hourly Data Sample:")
display(hourly_df.head())
```

```python
three_hour_df = pd.read_csv("../Albany_New_York_Weather/data/three_hour_data.csv")
print("\nThree-Hour Data Sample:")
display(three_hour_df.head())
```

```python
daily_df = pd.read_csv("../Albany_New_York_Weather/data/daily_data.csv")

print("\nDaily Data Sample:")
display(daily_df.head())
```

```python
monthly_df = pd.read_csv("../Albany_New_York_Weather/data/monthly_data.csv")
print("\nMonthly Data Sample:")
display(monthly_df.head())
```

```python
# Ensure 'DATE' is in datetime format
hourly_df['DATE'] = pd.to_datetime(hourly_df['DATE'], errors='coerce')

# Rename relevant column and convert to numeric
hourly_df.rename(columns={'HourlyDryBulbTemperature': 'Value'}, inplace=True)
hourly_df['Value'] = pd.to_numeric(hourly_df['Value'], errors='coerce')

# Fill missing values
hourly_df.fillna(hourly_df.mean(numeric_only=True), inplace=True)

# Display to verify
print("Transformed Hourly Data:")
display(hourly_df.head())

```

```python
three_hour_df['DATE'] = pd.to_datetime(three_hour_df['DATE'], errors='coerce')

three_hour_df.rename(columns={'HourlyDryBulbTemperature': 'Value'}, inplace=True)
three_hour_df['Value'] = pd.to_numeric(three_hour_df['Value'], errors='coerce')

three_hour_df.fillna(three_hour_df.mean(numeric_only=True), inplace=True)

print("Transformed Three-Hour Data:")
display(three_hour_df.head())
```

```python
daily_df['DATE'] = pd.to_datetime(daily_df['DATE'], errors='coerce').dt.date

daily_df.rename(columns={'DailyAverageDryBulbTemperature': 'Value'}, inplace=True)
daily_df['Value'] = pd.to_numeric(daily_df['Value'], errors='coerce')

daily_df.fillna(daily_df.mean(numeric_only=True), inplace=True)

print("Transformed Daily Data:")
display(daily_df.head())
```

```python
monthly_df['DATE'] = pd.to_datetime(monthly_df['DATE'], errors='coerce').dt.date

monthly_df.rename(columns={'MonthlyMeanTemperature': 'Value'}, inplace=True)
monthly_df['Value'] = pd.to_numeric(monthly_df['Value'], errors='coerce')

monthly_df.fillna(monthly_df.mean(numeric_only=True), inplace=True)

print("Transformed Monthly Data:")
display(monthly_df.head())
```

```python
# Aggregate hourly and three-hour data into daily averages
hourly_agg = hourly_df.groupby(hourly_df['DATE'].dt.date)['Value'].mean().reset_index()
three_hour_agg = three_hour_df.groupby(three_hour_df['DATE'].dt.date)['Value'].mean().reset_index()

# Merge the aggregated data
merged_data = pd.merge(hourly_agg, three_hour_agg, on='DATE', how='outer', suffixes=('_hourly', '_3hour'))

# Concatenate daily and monthly data for a summary
summary_data = pd.concat([daily_df[['DATE', 'Value']], monthly_df[['DATE', 'Value']]], ignore_index=True)
summary_data.rename(columns={'Value': 'SummaryValue'}, inplace=True)
```

```python
# Display the merged data and summary
print("Merged Hourly and Three-Hour Data:")
display(merged_data.head())
```

```python
# Display the merged data and summary
print("\nSummary Data (Daily and Monthly):")
display(summary_data.head())
```

```python
# Save the transformed data to new CSV files
merged_data.to_csv('data/merged_hourly_3hour_data.csv', index=False)
summary_data.to_csv('data/summary_daily_monthly_data.csv', index=False)

print("Transformed data has been saved to 'merged_hourly_3hour_data.csv' and 'summary_daily_monthly_data.csv'")
```
